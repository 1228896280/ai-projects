```python
# 基本csv读写操作
# 我们需要读取给定的训练数据，再进行后续的数据处理

def read_data(file_name):
    f = open(file_name)
    # ignore header
    f.readline()
    samples = []
    for line in f:
        line = line.strip().split(',')
        sample = [float(x) for x in line]
        samples.append(sample)
    return samples

def write_delimited_file(file_path, data, header=None, delimiter=','):
    f_out = open(file_path, 'w')
    if header is not None:
        f_out.write(delimiter.join(header) + '\n')
    for line in data:
        if isinstance(line, str):
            f_out.write(line + '\n')
        else:
            f_out.write(delimiter.join(line) + '\n')
    f_out.close()
```

# 2. 建模
我们这里简单的用sklearn中的逻辑回归来训练数据，建立模型，熟悉一下sklearn中的逻辑回归。

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
import math
import scipy

def lr_predict():
    # read in the training file
    train = read_data('./input/train.csv')
    print('读取训练数据完毕')
    # set the training response
    target = [x[0] for x in train]
    #set the training features
    train = [x[1:] for x in train]
    # read in the test file
    realtest = read_data('./input/test.csv')
    print('读取待预测数据')
    
    # code for logistic regression
    lr = LogisticRegression()
    lr.fit(train, target)
    print('Logistic Regression训练完毕!\n...\n')
    predicted_probs = lr.predict_proba(realtest)
    # write solutions to file
    predicted_probs = ['%f' % x[1] for x in predicted_probs]
    write_delimited_file('./output/lr_solution.csv', predicted_probs)
    print('Logistic Regression预测完毕! 请提交lr_solution.csv文件到Kaggle')

    
def svm_predict():
    train = read_data("./input/train.csv")
    print('读取训练数据完毕')
    target = [x[0] for x in train]
    train = [x[1:] for x in train]
    test = read_data("./input/test.csv")
    print('读取待预测数据')
    
    svc = svm.SVC(probability=True)
    svc.fit(train, target)
    print('SVM训练完毕!\n...\n')
    predicted_probs = svc.predict_proba(test)
    predicted_probs = ["%f" % x[1] for x in predicted_probs]
    write_delimited_file("./output/svm_solution.csv",
                                predicted_probs)
    print('SVM预测完毕! 请提交svm_solution.csv文件到Kaggle')
    
if __name__ == '__main__':
    svm_predict()
```
